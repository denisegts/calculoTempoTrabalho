```python
from dataclasses import dataclass
from typing import Dict, Tuple

# ============================================
# CONSTANTES
# ============================================
PROCESS_DURATIONS: Dict[str, float] = {
    'tipo_1': 5.91,    # minutos
    'tipo_2': 7.59,    # minutos
    'tipo_3': 0.85     # minutos
}

WORK_DAY_MINUTES: int = 510  # 8 horas e 30 minutos
WORK_DAY_HOURS: int = 8
WORK_DAY_EXTRA_MINUTES: int = 30


# ============================================
# CLASSES
# ============================================
@dataclass
class ResultadoTempo:
    """Armazena e formata os resultados do cálculo de tempo."""
    tempo_total_minutos: float
    tempo_tipo_1: float
    tempo_tipo_2: float
    tempo_tipo_3: float

    @property
    def horas(self) -> int:
        return int(self.tempo_total_minutos // 60)

    @property
    def minutos(self) -> int:
        return int(self.tempo_total_minutos % 60)

    @property
    def segundos(self) -> int:
        segundos_totais = int((self.tempo_total_minutos % 1) * 60)
        return segundos_totais

    @property
    def dentro_jornada(self) -> bool:
        return self.tempo_total_minutos <= WORK_DAY_MINUTES

    @property
    def diferenca_minutos(self) -> float:
        """Retorna tempo restante (negativo se exceder)."""
        return WORK_DAY_MINUTES - self.tempo_total_minutos
```

```python
# ============================================
# FUNÇÕES DE VALIDAÇÃO
# ============================================
def validar_entrada(num_tipo_1: int, num_tipo_2: int, num_tipo_3: int) -> Tuple[bool, str]:
    """
    Valida os valores de entrada.

    Args:
        num_tipo_1: Número de processos tipo 1
        num_tipo_2: Número de processos tipo 2
        num_tipo_3: Número de processos tipo 3

    Returns:
        Tupla (válido: bool, mensagem: str)
    """
    if num_tipo_1 < 0 or num_tipo_2 < 0 or num_tipo_3 < 0:
        return False, "❌ Erro: O número de processos não pode ser negativo!"

    if num_tipo_1 == 0 and num_tipo_2 == 0 and num_tipo_3 == 0:
        return False, "❌ Erro: Pelo menos um tipo de processo deve ser informado!"

    return True, "✅ Entrada validada com sucesso!"


# ============================================
# FUNÇÕES DE CÁLCULO
# ============================================
def calcular_tempo_trabalho(num_tipo_1: int, num_tipo_2: int, num_tipo_3: int) -> ResultadoTempo:
    """
    Calcula o tempo total de trabalho baseado no número de processos de cada tipo.

    Args:
        num_tipo_1: Número de processos de 5,91 minutos
        num_tipo_2: Número de processos de 7,59 minutos
        num_tipo_3: Número de processos de 0,85 minutos

    Returns:
        ResultadoTempo com todos os cálculos

    Raises:
        ValueError: Se os valores forem negativos
    """
    valido, mensagem = validar_entrada(num_tipo_1, num_tipo_2, num_tipo_3)
    if not valido:
        raise ValueError(mensagem)

    tempo_tipo_1 = num_tipo_1 * PROCESS_DURATIONS['tipo_1']
    tempo_tipo_2 = num_tipo_2 * PROCESS_DURATIONS['tipo_2']
    tempo_tipo_3 = num_tipo_3 * PROCESS_DURATIONS['tipo_3']
    tempo_total = tempo_tipo_1 + tempo_tipo_2 + tempo_tipo_3

    return ResultadoTempo(
        tempo_total_minutos=round(tempo_total, 2),
        tempo_tipo_1=round(tempo_tipo_1, 2),
        tempo_tipo_2=round(tempo_tipo_2, 2),
        tempo_tipo_3=round(tempo_tipo_3, 2)
    )
```

```python
# ============================================
# FUNÇÕES DE EXIBIÇÃO
# ============================================
def exibir_detalhamento_processos(resultado: ResultadoTempo, num_tipo_1: int, num_tipo_2: int, num_tipo_3: int) -> None:
    """
    Exibe detalhamento do tempo gasto em cada tipo de processo.

    Args:
        resultado: Objeto ResultadoTempo com os cálculos
        num_tipo_1: Quantidade de processos tipo 1
        num_tipo_2: Quantidade de processos tipo 2
        num_tipo_3: Quantidade de processos tipo 3
    """
    print("\n" + "="*60)
    print("📋 DETALHAMENTO POR TIPO DE PROCESSO")
    print("="*60)
    print(f"  Tipo 1 ({PROCESS_DURATIONS['tipo_1']} min/processo):")
    print(f"    • Quantidade: {num_tipo_1} processos")
    print(f"    • Tempo total: {resultado.tempo_tipo_1:.2f} minutos")

    print(f"\n  Tipo 2 ({PROCESS_DURATIONS['tipo_2']} min/processo):")
    print(f"    • Quantidade: {num_tipo_2} processos")
    print(f"    • Tempo total: {resultado.tempo_tipo_2:.2f} minutos")

    print(f"\n  Tipo 3 ({PROCESS_DURATIONS['tipo_3']} min/processo):")
    print(f"    • Quantidade: {num_tipo_3} processos")
    print(f"    • Tempo total: {resultado.tempo_tipo_3:.2f} minutos")
    print("="*60)


def exibir_resultado_final(resultado: ResultadoTempo) -> None:
    """
    Exibe o resultado final formatado com análise de jornada.

    Args:
        resultado: Objeto ResultadoTempo com os cálculos
    """
    print("\n" + "="*60)
    print("⏱️  RESULTADO FINAL")
    print("="*60)
    print(f"  Tempo total: {resultado.horas}h {resultado.minutos}m {resultado.segundos}s")
    print(f"  Em minutos: {resultado.tempo_total_minutos:.2f} min")
    print(f"\n  Jornada de trabalho: {WORK_DAY_HOURS}h {WORK_DAY_EXTRA_MINUTES}m ({WORK_DAY_MINUTES} min)")

    if resultado.dentro_jornada:
        diferenca = resultado.diferenca_minutos
        minutos_rest = int(diferenca)
        segundos_rest = int((diferenca % 1) * 60)
        print(f"\n  ✅ Status: DENTRO da jornada")
        print(f"  ⏳ Tempo restante: {minutos_rest}m {segundos_rest}s")
    else:
        excesso = abs(resultado.diferenca_minutos)
        minutos_exc = int(excesso)
        segundos_exc = int((excesso % 1) * 60)
        print(f"\n  ⚠️  Status: EXCEDE a jornada")
        print(f"  📌 Tempo em excesso: {minutos_exc}m {segundos_exc}s")

    print("="*60)


def processar_entrada_interativa() -> None:
    """
    Processa entrada interativa do usuário com tratamento de erros.
    """
    try:
        print("\n" + "="*60)
        print("📝 INSIRA OS DADOS")
        print("="*60)

        num_tipo_1 = int(input("Digite o número de processos tipo 1 (5,91 min): "))
        num_tipo_2 = int(input("Digite o número de processos tipo 2 (7,59 min): "))
        num_tipo_3 = int(input("Digite o número de processos tipo 3 (0,85 min): "))

        resultado = calcular_tempo_trabalho(num_tipo_1, num_tipo_2, num_tipo_3)
        exibir_detalhamento_processos(resultado, num_tipo_1, num_tipo_2, num_tipo_3)
        exibir_resultado_final(resultado)

    except ValueError as e:
        print(f"\n❌ Erro: {e}")
        print("Por favor, digite apenas números inteiros.")
    except Exception as e:
        print(f"\n❌ Erro inesperado: {e}")
```
